```python
# Import Library
import requests
from bs4 import BeautifulSoup

import time

def get_url(url):
    """Get parsed HTML from url
      Input: url to the webpage
      Output: Parsed HTML text of the webpage
    """
    print("RUN get_url, ", url)
    
    # Send GET request
    r = requests.get(url)

    # Parse HTML text
    soup = BeautifulSoup(r.text, 'html.parser')
    
    return soup

def cast_book(soup, base_url):
    # We want to save information about all products in a list
    print("RUN cast_book")
    data = []

    # Find all products
    products = soup.find_all('div', {'class':'product-item'})
    
    if products:
    # Extract information of each product
        for product in products:

            # Each product is dictionary containing the required information
            d = {'title':'', 'author':'', 'price_sale': 0, 'sale': 0, 'price_regular': 0, 'rating': 0, 'reviewers': 0, 'img_url':'', 'url':''}

            # We use the try-except blocks to handle errors
            try:
                # use strip for all text to remove outer blank space

                #get title and strip the space outside of it
                try:
                    title = product.find('p', {'class': 'title'})
                    if title:
                        d['title'] = title.text.strip()

                    #d['title'] = product['data-title'].strip()
                except:
                    print('Wrong with title')
                    print(product)

                try:
                    img_url = product.img['src']
                    if img_url:
                        d['img_url'] = img_url.strip()
                except:
                     print('Wrong with img_url')
                     print(product)

                try:
                    url = product.find("a")
                    if url:
                        d['url'] = base_url + url['href'].strip()
                except:
                    print('Wrong with url')
                    print(product)

                try:
                    author = product.find('p', {'class': 'author'})
                    if author:
                        d['author'] = product.find('p', {'class': 'author'}).text.strip()
                except:
                    print('Wrong with author')
                    print(product)

                #get price sale, strip outside space, split to get first item and remove the 'd' at the end
                #remove '.' in price sale to convert to int or it will make casting error
                try:
                    price_sale = product.find('span', {'class': 'final-price'})
                    if price_sale:
                        d['price_sale'] = int(price_sale.text.strip().split()[0][:-1].strip().replace('.',''))
                    # can get buy 'data-sale'
                except:
                    print('Wrong with price_sale')
                    print(product)

                try:
                    sale = product.find('span', {'class': 'sale-tag'})
                    if sale:
                        d['sale'] = int(sale.text.strip(' -%'))
                except:
                    print('Wrong with sale')
                    #print(product.find('span', {'class': 'sale-tag'}).text.strip())
                    print(product)

                try:
                    price_regular = product.find('span', {'class': 'price-regular'})
                    if price_regular:
                        d['price_regular'] = int(price_regular.text.strip()[:-1].replace('.',''))
                except:
                    print('Wrong with price_regular')
                    print(product)

                # get the rating, split by ':' and get the last, remove '%' 
                try:
                    rating_percentage = product.find('span', {'class': 'rating-content'}).span['style']
                    if rating_percentage:
                        rating = int(rating_percentage.split(':')[-1][:-1].strip())
                    # rating has 100% for 5 stars, 1 star is 20% so divide the rating for 20 
                    d['rating'] = round((rating) / 20, 1)
                except:
                    print('Wrong with rating')
                    print(product)

                try:
                    d['reviewers'] = int(product.find('p', {'class': 'review'}).text.strip('(').split()[0].strip())
                except:
                    print('Wrong with reviewers')
                    print(product)
                # Append the dictionary to data list
                data.append(d)

            except:
                # Skip if error and print error message
                print("We got error", product)      
        return data
    else:
        print("No products")
        return data

def scrape_tiki_by_category_and_page(base_url, category_url, page_url):
    """Scrape tiki by category book per page
    """
    #if else to change category_url 
    url = base_url + category_url + page_url
    print("RUN scrape_tiki_by_category_and_page")
    print(url)
    # Get parsed HTML
    soup = get_url(url)


    # Find all book
    data = cast_book(soup, base_url) 
#     print(data)
    return data
        
def scrape_tiki(time_to_sleep = 5):
    print("RUN scrape_tiki")
    data = []
    base_url = 'https://tiki.vn'
    category_url = '/nha-sach-tiki/c8322?src=c.8322.hamburger_menu_fly_out_banner'
    base_page_url = '&page='
    page = 1
    while page > 0:
        page_url = base_page_url + str(page)
        data_per_page = scrape_tiki_by_category_and_page(base_url, category_url, page_url)
        if len(data_per_page) == 0:
            break
        else:
            for i in data_per_page:
                data.append(i)
                
        time.sleep(10)
        page = page + 1
        print(page)
        
    return data
```
